```python



matriz_de_horarios_manha = [[None for _ in range(6)] for _ in range(5)]  # 5 horários de manhã
matriz_de_horarios_tarde = [[None for _ in range(6)] for _ in range(6)]  # 6 horários à tarde
matriz_de_horarios_noite = [[None for _ in range(6)] for _ in range(4)]  # 4 horários à noite

lista_horarios_manha = [" 08:00 - 08:55 ", " 08:55 - 09:50 ", " 10:00 - 10:55 ", " 10:55 - 11:50 ", " 12:00 - 12:55 "]
lista_horarios_tarde = [" 13:00 - 13:55 ", " 13:55 - 14:50 ", " 15:00 - 15:55 ", " 15:55 - 16:50 ", " 17:00 - 17:55 ", " 17:55 - 18:50 "]
lista_horarios_noite = [" 19:00 - 19:50 ", " 19:50 - 20:40 ", " 20:50 - 21:40 ", " 21:40 - 22:30 "]



def verificador_de_linhas_vazias(lista):
    for termo in lista:
        if termo is not None:
            return False
    return True


# Função para adicionar uma disciplina
def adicionar_disciplina(codigo_do_curso, dias, turno, horas):
    for hora in horas:
        hora_index = int(hora) - 1

        for dia in dias:
            dia_index = int(dia) - 2

            if turno == 'M':
                if codigo_do_curso == matriz_de_horarios_manha[hora_index][dia_index]:
                    continue
                matriz_de_horarios_manha[hora_index][dia_index] = codigo_do_curso

            elif turno == 'T':
                if codigo_do_curso == matriz_de_horarios_tarde[hora_index][dia_index]:
                    continue
                matriz_de_horarios_tarde[hora_index][dia_index] = codigo_do_curso

            elif turno == 'N':
                if codigo_do_curso == matriz_de_horarios_noite[hora_index][dia_index]:
                    continue
                matriz_de_horarios_noite[hora_index][dia_index] = codigo_do_curso
    return


# Função para remover uma disciplina
def remover_disciplina(codigo_do_curso, dias, turno, horas):
    for hora in horas:
        hora_index = int(hora) - 1

        for dia in dias:
            dia_index = int(dia) - 2

            if turno == 'M':
                if codigo_do_curso != matriz_de_horarios_manha[hora_index][dia_index]:
                    continue
                matriz_de_horarios_manha[hora_index][dia_index] = None

            elif turno == 'T':
                if codigo_do_curso != matriz_de_horarios_tarde[hora_index][dia_index]:
                    continue
                matriz_de_horarios_tarde[hora_index][dia_index] = None

            elif turno == 'N':
                if codigo_do_curso != matriz_de_horarios_noite[hora_index][dia_index]:
                    continue
                matriz_de_horarios_noite[hora_index][dia_index] = None
    return






def mostrar_horario():

    linha_tabela = ""
    codigo_do_curso = None

    # Formatação da saída conforme especificado
    tabela = "\n+---------------+----------+----------+----------+----------+----------+----------+\n" # lembrar que aqui apenas coloquei um \n para pular uma linha e não quebrar o inicio da tabela
    tabela += "|               | Seg      | Ter      | Qua      | Qui      | Sex      | Sab      |\n"
    tabela += "+---------------+----------+----------+----------+----------+----------+----------+\n"


    for index_horario in range(5):  # 5 horários - logo a matriz possui 5 linhas (5 listas)
        booleano_linha_vazia = verificador_de_linhas_vazias(matriz_de_horarios_manha[index_horario])
        if booleano_linha_vazia:
            continue

        linha_tabela = f"|{lista_horarios_manha[index_horario]}"

        for index_dia in range(6):  # 6 dias da semana - logo a matriz possui 6 colunas (6 termos em cada lista)
            codigo_do_curso = matriz_de_horarios_manha[index_horario][index_dia]
            if codigo_do_curso is None:
                linha_tabela += "|          "
            else:
                linha_tabela += f"| {codigo_do_curso} "
        linha_tabela += "|"
        tabela += linha_tabela + "\n"
        tabela += "+---------------+----------+----------+----------+----------+----------+----------+\n"



    for index_horario in range(6):  # 6 horários - logo a matriz possui 5 linhas (5 listas)

        booleano_linha_vazia = verificador_de_linhas_vazias(matriz_de_horarios_tarde[index_horario])
        if booleano_linha_vazia:
            continue

        linha_tabela = f"|{lista_horarios_tarde[index_horario]}"

        for index_dia in range(6):  # 6 dias da semana - logo a matriz possui 6 colunas (6 termos em cada lista)
            codigo_do_curso = matriz_de_horarios_tarde[index_horario][index_dia]
            if codigo_do_curso is None:
                linha_tabela += "|          "
            else:
                linha_tabela += f"| {codigo_do_curso} "
        linha_tabela += "|"
        tabela += linha_tabela + "\n"
        tabela += "+---------------+----------+----------+----------+----------+----------+----------+\n"



    for index_horario in range(4):  # 6 horários - logo a matriz possui 5 linhas (5 listas)

        booleano_linha_vazia = verificador_de_linhas_vazias(matriz_de_horarios_noite[index_horario])
        if booleano_linha_vazia:
            continue

        linha_tabela = f"|{lista_horarios_noite[index_horario]}"

        for index_dia in range(6):  # 6 dias da semana - logo a matriz possui 6 colunas (6 termos em cada lista)
            codigo_do_curso = matriz_de_horarios_noite[index_horario][index_dia]
            if codigo_do_curso is None:
                linha_tabela += "|          "
            else:
                linha_tabela += f"| {codigo_do_curso} "
        linha_tabela += "|"
        tabela += linha_tabela + "\n"
        tabela += "+---------------+----------+----------+----------+----------+----------+----------+\n"

    return tabela


def coletar_partes_da_instrucao(instrucao):
    partes_da_instrucao = instrucao.split()
    adicionar_ou_excluir = partes_da_instrucao[0]
    codigo_do_curso = partes_da_instrucao[1]
    dths = partes_da_instrucao[2:]

    matriz_dths = []
    lista_dias = []
    lista_horarios = []

    for dth in dths:
        for i in range(len(dth)): # tratamento da string!!!!!!!!
            if dth[i] in "MTN":
                turno = dth[i]
                dias = dth[:i]
                horarios = dth[i+1:]
                break

        for dia in dias:
            lista_dias.append(dia)

        for horario in horarios:
            lista_horarios.append(horario)

        matriz_dths.append([lista_dias, turno, lista_horarios])

    return adicionar_ou_excluir, codigo_do_curso, matriz_dths


def main():
    while True:
        instrucao = input()
        if instrucao == "Hasta la vista, beibe!":
            print("\n")
            break

        elif instrucao == "?":
            print(mostrar_horario())

        else:

            partes_da_instrucao = coletar_partes_da_instrucao(instrucao)

            adicionar_ou_excluir = partes_da_instrucao[0]
            codigo_do_curso = partes_da_instrucao[1]
            matriz_dths = partes_da_instrucao[2]

            for dth in matriz_dths:
                dias = dth[0]
                turno = dth[1]
                horas = dth[2]

                if adicionar_ou_excluir == "+":
                    adicionar_disciplina(codigo_do_curso, dias, turno, horas)

                elif adicionar_ou_excluir == "-":
                    remover_disciplina(codigo_do_curso, dias, turno, horas)


main()
```
